# Dataset analytics for SRS

This notebook is used to create a statistically representative subset from the years 2020-2022.

```python
import sys, os
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Add the MscThesis directory to the Python path
sys.path.append(os.path.abspath(os.path.join('..')))

from Utils.constants import COMBINED_PATH, METADATA_PATH
```

```python
# Read individual CSVs
df1 = pd.read_csv("C:/Users/Chloe/Documents/MaastrichtLaw&Tech/Thesis/MscThesis/Dataset/Metadata/2020_rs_data.csv")
df2 = pd.read_csv("C:/Users/Chloe/Documents/MaastrichtLaw&Tech/Thesis/MscThesis/Dataset/Metadata/2021_rs_data.csv")
df3 = pd.read_csv("C:/Users/Chloe/Documents/MaastrichtLaw&Tech/Thesis/MscThesis/Dataset/Metadata/2022_rs_data.csv")

# Concatenate dataframes vertically
combined = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
combined.head()
```

```python
combined['year'] = pd.to_datetime(combined['date']).dt.year
combined.to_csv("C:/Users/Chloe/Documents/MaastrichtLaw&Tech/Thesis/MscThesis/Dataset/Metadata/combined.csv", index=False)
combined.head()
```

```python
# Group by 'year' and 'instantie', then count the occurrences
grouped = combined.groupby(['year', 'instantie']).size().reset_index(name='value')
```

```python
# Get unique years and instanties
years = grouped['year'].unique()
instanties = grouped['instantie'].unique()
```

```python
grouped
```

```python
# Create an empty dictionary to hold the base heights for each year
base_height = {year: 0 for year in years}

# Set the figure size
plt.figure(figsize=(10, 10))
```

```python
# Loop through each 'instantie' and create the stacked bars
for instantie in instanties:
    # Get the subset of data for this 'instantie'
    subset = grouped[grouped['instantie'] == instantie]

    # Plot the bars for this 'instantie'
    plt.bar(
        subset['year'],
        subset['value'],
        bottom=[base_height[year] for year in subset['year']],
        label=instantie
    )

    # Update the base height for each year
    for i, row in subset.iterrows():
        base_height[row['year']] += row['value']

```

```python
# Set labels and legend
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Stacked Bar Graph with Matplotlib')
plt.legend(title='Instantie')
plt.show()
```

```python
plt.savefig('Stats/stats_full_20_22.png')
plt.close()
```

# Create subset

```python
sampling_ratio = 0.05

groups = grouped['instantie'].tolist()
values = grouped['value'].tolist()

combined_list = list(zip(groups, values))

years = [2020, 2021, 2022]

print(combined_list)
```

```python
print(len(combined))
sampled_entries = []
for _,row in grouped.iterrows():
    nr_samples = int(sampling_ratio * row['value'])
    # Correct boolean conditions with parentheses
    condition = (combined['instantie'] == row['instantie']) & (combined['year'] == row['year'])

    # Keep part of the DataFrame meeting the condition
    entries_to_sample = combined[condition]

    entries_to_sample = entries_to_sample.sample(n=nr_samples, random_state=42)
    sampled_entries.extend(entries_to_sample.ecli)

print(sampled_entries)
print(len(sampled_entries))
```

```python
sample_set = combined[combined['ecli'].isin(sampled_entries).reset_index(drop=True)]
sample_set
```

```python
sample_set.to_csv("C:/Users/Chloe/Documents/MaastrichtLaw&Tech/Thesis/MscThesis/Dataset/Metadata/subset.csv")
```
